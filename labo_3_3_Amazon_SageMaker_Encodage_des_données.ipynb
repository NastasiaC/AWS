# **Deuxième étape :** ENCODAGE DES CARACTÉRISTIQUES ORDINALES

```python
df_car.info()
```

```python
#total des valeurs de la colonne nombres de portes
df_car['num-of-doors'].value_counts()
```

```python
#création d'un dictionnaire two est associé au chiffre 2 et four est associé au chiffre 4
#le dictionnaire est concu pour assoicer des répresentation textuelle de nombres à leur équivalence numérique
door_mapper = {"two": 2,
              "four": 4}
```

```python
#ajout colonne "doors" dans le fichier "df_car"
#sélection de la colonne existante "num-of-doors"
#remplacer les valeurs dans la colonne sélectionner la colonne door_mapper est le dictionnaire défini précédemment
df_car['doors'] = df_car["num-of-doors"].replace(door_mapper)
```

```python
#affichage des 5 premières lignes
df_car.head()
```

```python
#affichage du total des valeurs de la colonne "nombres de cylindres"
df_car['num-of-cylinders'].value_counts()
```

```python
#création d'un dictionnaire le dictionnaire est concu pour assoicer des répresentation textuelle de nombres à leur équivalence numérique
cylinder_mapper = {"two":2,
                  "three":3,
                  "four":4,
                  "five":5,
                  "six":6,
                  "eight":8,
                  "twelve":12}
```

```python
#ajout colonne "cylinder" dans le fichier "df_car"
#sélection de la colonne existante "num-of-cylinders"
#remplacer les valeurs dans la colonne sélectionner la colonne door_mapper est le dictionnaire défini précédemment
df_car['cylinders'] = df_car['num-of-cylinders'].replace(cylinder_mapper)
```

```python
#affichage de 5 premièr lignes du tableau
df_car.head()
```

# **Troisième étape :** ENCODAGE DES CARACTÉRISTIQUES NON-ORDIANLES

```python
#affichage des valeurs totales de la colonne roues motrices
df_car['drive-wheels'].value_counts()
```

```python
df_car = pd.get_dummies(df_car,columns=['drive-wheels'])
```

```python
#affichage des 5 premières lignes du tableau
df_car.head()
```

```python
#affichage des valeurs totales de la colonne aspirations
df_car['aspiration'].value_counts()
```

```python
#pd.get_dummies : est-um fonction dans pandas qui permet des variables indicatrices à partir de varaibles catégorielles
#drop_first=True : option qui supprime la première catégorie encodée pour éviter la multicoréalité
#La multicolinéarité se produit lorsque deux ou plusieurs variables explicatives (indépendantes) dans un modèle de régression sont fortement corrélées entre elles
df_car = pd.get_dummies(df_car,columns=['aspiration'], drop_first=True)
```
