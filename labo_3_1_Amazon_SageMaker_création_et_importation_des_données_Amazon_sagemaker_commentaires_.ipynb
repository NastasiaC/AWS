# **IMPORTATION DES DONNÉES**

```python
# Modules Warning : permet de gérer des messages d'avertissements dans Python cela va permettre d'informer l'utilisateur de comportement non optimal dans le code
# Modules requests : bibliothèque populaire pour effectuer des requêtes HTTP de manière simple et lisible
# Modules zipfile : lire et écrire des fichiers compressés dans un format ZIP, cela permet d'extraire le contenu d'un fichier zip ou créer un nouveau fichier zip
# Module IO : fournit des outils pour manipuler les flux d'entrées/sorties en mémoire, particulièrement pour travailler avec des données texte ou en binaire, permet de créer des flux de données en mémoire sans écrire des fichiers physiques
    import warnings, requests, zipfile, io
# permet d'ignorer tous les avertissements dans un script Python, évite les messages d'avertissements indésirables
    warnings.simplefilter('ignore')
# Module pandas : manipulation et analyse des données, utilisation dans les domaines science des données, finance, analyse des données et machine learning, cela va générer des données structurée et facile à utiliser
    import pandas as pd
#chargement des données au format arff en utilisant scipy
#Scipy : permet la lecture de fichier ARFF et convertir en structure de donnée utilisable
#ARFF : format de fichier utilisé pour stocker des ensembles de donnée. Ce format a été créé pour l'outil de machine learning - ARFF est similaire au format CSV mais contient plus d'informations et types d'attributs
    from scipy.io import arff
```

```python
    #fichier zip
    f_zip = 'http://archive.ics.uci.edu/ml/machine-learning-databases/00212/vertebral_column_data.zip'
    #Téléchargement des données par morceaux et non d'un coup
    # requests.get : moyen pratique d'envoyer des requêtes HTTP GET en python en utilisant Requests
    # stream=True : permet de télécharger des fichiers sans charger tout leur contenu en mémoire d'un coup
    r = requests.get(f_zip, stream=True)
    Vertebral_zip = zipfile.ZipFile(io.BytesIO(r.content))
    Vertebral_zip.extractall()
```
